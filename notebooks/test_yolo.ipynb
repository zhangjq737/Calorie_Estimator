```python
# Cell 1: Imports
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
# Cell 2: Load pre-trained YOLOv11
model = YOLO('../models/yolo11n.pt')

print("Model loaded successfully!")
print(f"Device: {model.device}")
print(f"Classes: {model.names}")
```

```python
# Cell 3: Test on sample image
# img_url = "https://ultralytics.com/images/zidane.jpg"
img_path = "../data/mixed_plate2.jpg"

results = model(img_path)
```

```python
# Cell 4: Visualize results
result = results[0]

# Or plot in notebook
img_with_boxes = result.plot()
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Cell 5: Check detected classes
for box in result.boxes:
    class_id = int(box.cls[0])
    class_name = result.names[class_id]
    confidence = float(box.conf[0])
    print(f"{class_id} Detected: {class_name} (confidence: {confidence:.2f})")
```
